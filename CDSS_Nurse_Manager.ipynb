```python
import pandas as pd

df = pd.read_csv("nurse_dataset.csv")

overtime_risk = df[df['weekly_hours'] > 40]
print("--- Staff at Risk of Burnout ---")
print(overtime_risk[['name' , 'department' , 'weekly_hours']])

```

```python


edu_distribution = df.groupby('department')['education_level'].value_counts().unstack()
print("\n--- Education Level Distribution per Department ---")
print(edu_distribution)
```

```python
eligiable_candidates = df[
    (df['department'] == 'ICU') &
    (df['education_level'].str.contains('Master', na=False))
&
    (df['weekly_hours'] < 35)
]

print("\n--- Eligible Candidates for ICU Department ---")
print(eligiable_candidates)
```
